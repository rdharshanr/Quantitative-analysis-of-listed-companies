### **Ratio Analysis**

## **Liquidity ratios**

```python
Total_Current_asset_HDFC=HDFC_BS.loc['Current Assets'].fillna(0)+HDFC_BS.loc['Other Current Assets'].fillna(0)+HDFC_BS.loc['Restricted Cash'].fillna(0)+HDFC_BS.loc['Prepaid Assets'].fillna(0)+HDFC_BS.loc[ 'Other Receivables'].fillna(0)+HDFC_BS.loc['Taxes Receivable'].fillna(0)+HDFC_BS.loc['Accounts Receivable'].fillna(0)+HDFC_BS.loc['Cash Cash Equivalents And Short Term Investments'].fillna(0)	+HDFC_BS.loc['Other Short Term Investments'].fillna(0)
+HDFC_BS.loc['Cash And Cash Equivalents'].fillna(0)+HDFC_BS.loc['Cash Equivalents'].fillna(0)+HDFC_BS.loc['Cash Financial'].fillna(0)
Total_Current_asset_HDFC.astype(int)
```

```python
Total_Current_Liabilities_HDFC = (
    HDFC_BS.loc['Other Current Liabilities'].fillna(0) +
    HDFC_BS.loc['Current Debt And Capital Lease Obligation'].fillna(0) +
    HDFC_BS.loc['Current Capital Lease Obligation'].fillna(0) +
    HDFC_BS.loc['Pensionand Other Post Retirement Benefit Plans Current'].fillna(0) +
    HDFC_BS.loc['Payables'].fillna(0) +
    HDFC_BS.loc['Other Payable'].fillna(0) +
    HDFC_BS.loc['Total Tax Payable'].fillna(0) +
    HDFC_BS.loc['Accounts Payable'].fillna(0)
)
```

```python
Total_Current_Liabilities_HDFC.astype(int)
```

```python
print('Liquidity ratios')
print('-'*30)
print('Current Ratio')
# current ratio = Total Current Assets/Total Current Liabilities
HDFC_CR = Total_Current_asset_HDFC / Total_Current_Liabilities_HDFC # Total Current Assets/Total Current Liabilities
print(HDFC_CR,"\n")
print('Quick Ratio')
#Quick ratio = (Total Current Assets-Inventory)/Total Current Liabilities, no iventory
HDFC_QR =[Total_Current_asset_HDFC / Total_Current_Liabilities_HDFC.astype(int)]
print(HDFC_QR,"\n")


```

```python
plt.figure(figsize=(15, 6))
plt.plot(HDFC_CR.index, HDFC_CR, marker='o', color='blue')

plt.title('HDFC Current Ratio Over Time')
plt.xlabel('Financial Year')
plt.ylabel('Current Ratio')
plt.grid(True)
plt.show()


```

### **Asset Utilization or Turnover ratios**

```python
print('Total Asset Turnover')
TAT_HDFC = HDFC_Financials.loc['Total Revenue'].astype(int)/HDFC_BS.loc['Total Assets'].astype(int)# Total Revenue / Total Assets
print(TAT_HDFC,"\n")
print('Average Collection Period')
ACP_HDFC= HDFC_BS.loc['Accounts Receivable'].astype(int) / (HDFC_Financials.loc['Total Revenue'].astype(int)/365) # Net Receivables/(Total Revenue/360)
print(ACP_HDFC,"\n")
```

### **Leverage ratios**

```python
print('Financial Leverage ratios')
print('-'*40)
print('Total Debt Ratio')
HDFC_TDR = HDFC_BS.loc['Total Debt'].astype(int) / HDFC_BS.loc['Total Assets'].astype(int) # Total Liabilities / Total Assets
print(HDFC_TDR,"\n")
print('Debt to equity ratio')
HDFC_DE = HDFC_BS.loc['Total Debt'].astype(int) / HDFC_BS.loc["Stockholders Equity"].astype(int) # Total Liabilities / Total stockholders' equity
print(HDFC_DE,"\n")
```

### **Profitability ratios**

```python
print('Profitability ratios')
print('-'*20)
print('Gross Profit Margin')
HDFC_GPM = HDFC_Financials.loc['Gross Profit'].astype(int) / HDFC_Financials.loc['Total Revenue'].astype(int) # Gross Profit / Total Revenue
print(HDFC_GPM,"\n")
print('Net Profit Margin')
HDFC_NPM = HDFC_Financials.loc['Net Income Common Stockholders'].astype(int) / HDFC_Financials.loc['Total Revenue'].astype(int) # Net Income / Total Revenue
print(HDFC_NPM,"\n")
print('Return on Assets (ROA)')
HDFC_ROA = HDFC_Financials.loc['Net Income Common Stockholders'].astype(int) / HDFC_BS.loc['Total Assets'].astype(int) # Net Income / Total Assets
print(HDFC_ROA,"\n")
print('Return on Equity (ROE)')
HDFC_ROE = HDFC_Financials.loc['Net Income Common Stockholders'].astype(int) / HDFC_BS.loc['Stockholders Equity'].astype(int) # Net Income / Total Equity
print(HDFC_ROE,"\n")
print('Earning Per Share (EPS)')
HDFC_EPS = HDFC_Financials.loc['Net Income Common Stockholders'].astype(int) / HDFC_BS.loc['Share Issued'].astype(int)  # Net Income / (Share issued)
print(HDFC_EPS,"\n")
```

```python
plt.figure(figsize=(12, 8))

# Define financial years
years = HDFC_GPM.index.astype(str)

# Set up positions for bars
bar_width = 0.15
bar_positions = np.arange(len(years))

# Plotting Gross Profit Margin
plt.bar(bar_positions - 2 * bar_width, HDFC_GPM, width=bar_width, label='Gross Profit Margin', color='blue')

# Plotting Net Profit Margin
plt.bar(bar_positions - bar_width, HDFC_NPM, width=bar_width, label='Net Profit Margin', color='orange')

# Plotting Return on Assets (ROA)
plt.bar(bar_positions, HDFC_ROA, width=bar_width, label='ROA', color='green')

# Plotting Return on Equity (ROE)
plt.bar(bar_positions + bar_width, HDFC_ROE, width=bar_width, label='ROE', color='red')



# Adding annotations
for i, value in enumerate(HDFC_GPM):
    plt.text(i - 2 * bar_width, value + 0.01, f'{value:.2f}', ha='center', va='bottom', fontsize=8)

for i, value in enumerate(HDFC_NPM):
    plt.text(i - bar_width, value + 0.01, f'{value:.2f}', ha='center', va='bottom', fontsize=8)

for i, value in enumerate(HDFC_ROA):
    plt.text(i, value + 0.01, f'{value:.2f}', ha='center', va='bottom', fontsize=8)

for i, value in enumerate(HDFC_ROE):
    plt.text(i + bar_width, value + 0.01, f'{value:.2f}', ha='center', va='bottom', fontsize=8)



# Adding labels and legend
plt.xlabel('Financial Year')
plt.ylabel('Ratios')
plt.title('Profitability Ratios Over Time')
plt.xticks(bar_positions, years)
plt.legend()

plt.tight_layout()
plt.show()
```
